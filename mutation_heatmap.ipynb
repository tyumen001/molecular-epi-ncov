# （三）Melt-ba5/bf7/xbb

## 1.delete ancestral mutations

```python
import pandas as pd
for file_name in {'china_ba5', 'china_bf7', 'china_xbb', 'world_xbb'}:
    mutation = pd.read_csv(f'8.melt/{file_name}_mutation.tsv', sep='\t')
    # anceatral mutations
    ancestral_mutation = pd.read_csv(f'1.ancestral/{file_name}_ancestral_mutation.tsv', sep='\t')

    mutation = mutation[~mutation['Mutation'].isin(ancestral_mutation['Mutation'])]
    mutation[['Mutation', 'Count']].to_csv(f'8.melt/{file_name}_mut_unfix.tsv', sep='\t', index=False)
```

## 2.Handling ancestral node mutations at the same site

```python
# For example: Modify ancestor T478K and node T478R to K478R

for file_name in {'china_ba5', 'china_bf7', 'china_xbb', 'world_xbb'}:
    mut_ancestral = pd.read_csv(f'1.ancestral/{file_name}_ancestral_mutation.tsv', sep='\t')
    mut_seq = pd.read_csv(f'8.melt/{file_name}_mut_unfix.tsv', sep='\t')

    # split
    mut_seq['single_Mutation'] = mut_seq['Mutation']
    mut_seq[['Gene', 'single_Mutation']] = mut_seq['single_Mutation'].str.split(':', expand=True)
    mut_seq[['Original_AA', 'Position', 'Mutated_AA']] = mut_seq['single_Mutation'].str.extract('(\D)(\d+)(\D)')

    # copy
    fix_mut_seq = mut_seq.copy()

    mut_ancestral['single_Mutation'] = mut_ancestral['Mutation']
    mut_ancestral[['Gene', 'single_Mutation']] = mut_ancestral['single_Mutation'].str.split(':', expand=True)
    mut_ancestral[['Original_AA', 'Position', 'Mutated_AA']] = mut_ancestral['single_Mutation'].str.extract('(\D)(\d+)(\D)')

    # Find rows in mut_seq and mut_ancestral where Gene, Original_AA, and Position columns are the same but Mutated_AA is different
    for index1, row1 in mut_ancestral.iterrows():
        # modify
        for index2, row2 in fix_mut_seq.iterrows():
            matching_row = (fix_mut_seq.iloc[index2, 3] == mut_ancestral.iloc[index1, 3]) & \
                        (fix_mut_seq.iloc[index2, 4] == mut_ancestral.iloc[index1, 4]) & \
                        (fix_mut_seq.iloc[index2, 5] == mut_ancestral.iloc[index1, 5]) & \
                        (fix_mut_seq.iloc[index2, 6] != mut_ancestral.iloc[index1, 6])
            if matching_row:
                print("ancestral:" + mut_ancestral.iloc[index1, 0] + "    seq:" + fix_mut_seq.iloc[index2, 0])
                # replace
                fix_mut = fix_mut_seq.iloc[index2, 3] + ":" + mut_ancestral.iloc[index1, 6] + fix_mut_seq.iloc[index2, 5] + fix_mut_seq.iloc[index2, 6]
                print(fix_mut)
                fix_mut_seq.iloc[index2, 0] = fix_mut

    fix_mut_seq[['Mutation', 'Count']].to_csv(f'8.melt/{file_name}_mut.tsv', sep='\t', index=False)
```

## 3.Count frequency and filter（3% < f < 100%）

```python
# melt data
import pandas as pd

# number of sequences
china_ba5_seq_num = 11076
china_bf7_seq_num = 5290
china_xbb_seq_num = 18266
world_xbb_seq_num = 349829

china_ba5 = pd.read_csv('8.melt/china_ba5_mut.tsv', sep='\t')
china_bf7 = pd.read_csv('8.melt/china_bf7_mut.tsv', sep='\t')
china_xbb = pd.read_csv('8.melt/china_xbb_mut.tsv', sep='\t')
world_xbb = pd.read_csv('8.melt/world_xbb_mut.tsv', sep='\t')

china_ba5['Frequency_china'] = china_ba5['Count'] / china_ba5_seq_num
china_bf7['Frequency_china'] = china_bf7['Count'] / china_bf7_seq_num
china_xbb['Frequency_china'] = china_xbb['Count'] / china_xbb_seq_num
world_xbb['Frequency_world'] = world_xbb['Count'] / world_xbb_seq_num

china_world_mutation = pd.merge(china_ba5, china_bf7, on='Mutation', how='outer')
china_world_mutation.rename(columns={'Count_x': 'Count_china_ba5', 
                                     'Frequency_china_x': 'Frequency_china_ba5', 
                                     'Count_y': 'Count_china_bf7', 
                                     'Frequency_china_y': 'Frequency_china_bf7'}, 
                                     inplace=True)
china_world_mutation = pd.merge(china_world_mutation, china_xbb, on='Mutation', how='outer')
china_world_mutation = pd.merge(china_world_mutation, world_xbb, on='Mutation', how='outer')
china_world_mutation.rename(columns={'Count_x': 'Count_china_xbb', 
                                     'Frequency_china': 'Frequency_china_xbb', 
                                     'Count_y': 'Count_world_xbb', 
                                     'Frequency_world': 'Frequency_world_xbb'}, 
                                     inplace=True)

china_world_mutation_filter = china_world_mutation[(china_world_mutation['Frequency_china_ba5'].between(0.03, 1) |
                                                   china_world_mutation['Frequency_china_bf7'].between(0.03, 1) |
                                                   china_world_mutation['Frequency_china_xbb'].between(0.03, 1) |
                                                   china_world_mutation['Frequency_world_xbb'].between(0.03, 1))]

china_world_mutation_filter.to_csv('8.melt/china_world_mutation2.tsv', sep='\t', index=False)
```

```python
# mark gene
import pandas as pd

df = pd.read_csv('8.melt/china_world_mutation2.tsv', sep='\t')
df[['Gene', 'Mutation']] = df['Mutation'].str.split(':', expand=True)
df[['Original_AA', 'Position', 'Mutated_AA']] = df['Mutation'].str.extract('(\D)(\d+)(\D)')
df = df[['Mutation', 'Frequency_china_ba5', 'Frequency_china_bf7', 'Frequency_china_xbb', 'Frequency_world_xbb', 'Gene', 'Original_AA', 'Position', 'Mutated_AA']]
df.to_csv('8.melt/china_world_mutation_frequency.tsv', sep='\t', index=False)
```

## 4.Make a figure


```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
import pandas as pd

rcParams['font.family'] = 'Times New Roman'

df = pd.read_csv('8.melt/china_world_mutation_frequency.tsv', sep='\t')

for gene in df['Gene'].unique():
    df_gene = df[df['Gene'] == gene]
    # x.axis.name
    df_gene.rename(columns={'Frequency_china_ba5': 'China BA.5.2.48', 
                            'Frequency_china_bf7': 'China BF.7', 
                            'Frequency_china_xbb': 'China XBB', 
                            'Frequency_world_xbb': 'World XBB', 
                            }, inplace=True)

    # normalize by row
    df_gene_norm = df_gene.fillna(0)
    # Mark larger values and end up showing >0.3
    df_gene_norm.iloc[:, 1:5] = df_gene_norm.iloc[:, 1:5].apply(lambda x: x.map(lambda y: 0.3 if y > 0.3 else y))

    # standardization
    #df_s_gene_norm.iloc[:, 1:4] = df_s_gene_norm.iloc[:, 1:4].apply(lambda x: (x - x.mean()) / x.std(), axis=1) # row
    #df_s_gene_norm.iloc[:, 1:4] = df_s_gene_norm.iloc[:, 1:4].apply(lambda x: (x - x.mean()) / x.std(), axis=0) # column

    # sort
    df_gene_norm = df_gene_norm.sort_values('Position')

    # plot
    #plt.figure(figsize=(2, 5.5))
    if len(df_gene_norm) == 15:
        height = 5.5
    elif len(df_gene_norm) == 14:
        height = 5
    elif len(df_gene_norm) == 13:
        height = 5
    elif len(df_gene_norm) == 5:
        height = 2
    elif len(df_gene_norm) == 7:
        height = 2.5
    elif len(df_gene_norm) <= 5:
        height = 1.5
    
    plt.figure(figsize=(2, height))
    
    sns.heatmap(df_gene_norm[['Mutation', 'China BA.5.2.48', 'China BF.7', 'China XBB', 'World XBB']].set_index('Mutation'), 
                #annot=True,
                linewidths=2, linecolor='grey', cmap='YlGnBu', vmin=0, vmax=0.3)
    plt.yticks(rotation=0)  # Set x-axis tick labels to be horizontal
    #plt.title('Heatmap of Mutation Frequency for S Gene')
    plt.title('Mutation heatmap for ' + gene +' Gene')
    
    plt.savefig('plot/' + gene + '-heatmap.png', dpi=300, bbox_inches='tight')
```
